Cleaning

```python
def final_clean(text):
  clean_text = re.compile('https?://\S+|www\.\S+|<.*?>|[^a-zA-Z\s]|\d+')
  final_text = re.sub(clean_text,'',text)
  return final_text
```

```python
df['text'] = df['text'].apply(final_clean)
```

Tokenization,Stopwords and Lemitization

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
```

```python
def TSL(text):
  T = word_tokenize(text)
  stop_words = set(stopwords.words('english'))
  filtered = []
  for w in T:
    if w not in stop_words:
      filtered.append(w)
  lemma = WordNetLemmatizer()
  for i in range(len(filtered)):
    filtered[i] = lemma.lemmatize(filtered[i])
  new_text = ' '.join(filtered)
  return new_text
```

```python
df['text'] = df['text'].apply(TSL)
```
